```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather_df = pd.read_csv('weather_4provinces_daily_2021_2025.csv')
weather_df.head()
```

```python
weather_wide = weather_df.pivot(
    index='Date',
    columns='Province',
    values=['Temperature_C', 'Rainfall_mm', 'Humidity_Percent', 'Soil_Moisture']
)

new_names = []
for col in weather_wide.columns:
    variable_name = col[0]
    province_name = col[1]

    full_name = f"{province_name}_{variable_name}"

    new_names.append(full_name)

weather_wide.columns = new_names

weather_wide.head()
```

```python
weather_wide = weather_wide.reset_index()
weather_wide['Date'] = weather_wide['Date'].astype(str)

temp_split = weather_wide['Date'].str.split('/', expand=True)

weather_wide['Day'] = temp_split[0].astype(int)
weather_wide['Month'] = temp_split[1].astype(int)
weather_wide['Year'] = temp_split[2].astype(int)

temp_date = pd.to_datetime(weather_wide[['Year', 'Month', 'Day']])
weather_wide['DayOfWeek_Num'] = temp_date.dt.dayofweek
weather_wide.head()

```

```python
weather_wide.to_csv('weather_data.csv', index=False)
```

Diesel

```python
df_oil2564 = pd.read_csv('./dataset/oil_2564.csv', header=1)
df_oil2565 = pd.read_csv('./dataset/oil_2565.csv', header=1)
df_oil2566 = pd.read_csv('./dataset/oil_2566.csv', header=1)
df_oil2567 = pd.read_csv('./dataset/oil_2567.csv', header=1)
df_oil2568 = pd.read_csv('./dataset/oil_2568.csv', header=1)
```

```python
diesel_data = pd.concat([df_oil2564, df_oil2565, df_oil2566, df_oil2567, df_oil2568], ignore_index=True)
diesel_data = diesel_data.rename(columns={'Unnamed: 0': 'Date', 'ไฮดีเซล': 'Diesel'})
diesel_data = diesel_data[['Date', 'Diesel']]
diesel_data.head()
```

```python
diesel_data['Date'] = diesel_data['Date'].astype(str)

temp_split = diesel_data['Date'].str.split('/', expand=True)

diesel_data['Day'] = temp_split[0].astype(int)
diesel_data['Month'] = temp_split[1].astype(int)
diesel_data['Year'] = temp_split[2].astype(int)

diesel_data['Year'] = diesel_data['Year'] - 543


diesel_data.head()
```

```python
temp_date = pd.to_datetime(diesel_data[['Year', 'Month', 'Day']])
diesel_data['DayOfWeek_Num'] = temp_date.dt.dayofweek
# diesel_data = diesel_data.drop(columns=['Date'])
diesel_data.head()
```

```python
diesel_data.to_csv('cleaned_oil_data.csv', index=False)
```

Ingredient dataset cleaning

```python
merge02_data = pd.read_csv('merge02_data.csv')
merge02_data['Date_Match'] = pd.to_datetime(merge02_data['Date_Match'])

files = ["chicken", "chili", "coriander", "egg", "garlic", "lime", "onion", "peanut", "pork",
         "shallot", "shrimp", "spring_onion", "tamarind_juice"]

for i in files:    

    loop_data = pd.read_csv(f'./dataset/ingredient/{i}.csv')

    loop_data['วันที่'] = loop_data['วันที่'].replace('29 ก.พ. 2567', '29/2/2567')
    loop_data = loop_data.rename(columns={'ราคาเฉลี่ย': f'Price_{i}'})
    

    loop_data['Date'] = loop_data['วันที่'].astype(str)

    temp_split = loop_data['Date'].str.split('/', expand=True)

    loop_data['Day'] = temp_split[0].astype(int)
    loop_data['Month'] = temp_split[1].astype(int)
    loop_data['Year'] = temp_split[2].astype(int) - 543
    

    loop_data['Date_Match'] = pd.to_datetime(loop_data[['Year', 'Month', 'Day']])

    loop_data_clean = loop_data[[f'Price_{i}', 'Date_Match']]

    merge02_data = pd.merge(
                merge02_data, 
                loop_data_clean, 
                on='Date_Match', 
                how='left'
                )
```

```python
merge02_data.head()
merge02_data.to_csv('merge03_df.csv', index=False)
```

```python
names =  ['Price_chicken', 'Price_chili', 'Price_coriander', 'Price_egg', 'Price_garlic', 'Price_lime',
       'Price_onion', 'Price_peanut', 'Price_pork', 'Price_shallot',
       'Price_shrimp', 'Price_spring_onion', 'Price_tamarind_juice']

ingredient_df = ingredient_df.sort_values(by='Date_Match')

for i in names:
    

    ingredient_df[i] = ingredient_df[i].ffill()
    ingredient_df[i] = ingredient_df[i].bfill()


ingredient_df.head()
```
